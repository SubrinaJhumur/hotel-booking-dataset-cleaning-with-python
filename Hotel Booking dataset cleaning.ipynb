```python
# the link of the dtaaset: https://www.kaggle.com/datasets/jessemostipak/hotel-booking-demand
# import pandas for data loading and data cleaning
import pandas as pd
```

```python
# checking the duplicate values present in the dataset
data.duplicated().sum()
```

```python
# removing the duplicate values from the dataset
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
# check the unique values in the agent column
data['agent'].unique()
data[data['agent']==5]
```

```python
# filling the na values of the agent column with -1 values
data['agent'].fillna(-1,inplace=True)
```

```python
# drop the missing values present in the 'num_of_child' column
data.dropna(subset=['num_of_child'],inplace=True)
# missing values in the country column is filled with unknown word
data['country'].fillna('unknown',inplace=True)
# check is there any null values present in the dataset
data.isnull().sum()
```

```python
# check the data types
data.dtypes
```

```python
data.columns
```

```python
# converting the 'is_canceled' and 'is_repeated_guest' column into boolean values and 'num_of_child' and 'agent' column into int values
data=data.astype({'is_canceled':'boolean','is_repeated_guest':'boolean','num_of_child':'int','agent':'int'})
```

```python
# after converting the data types
data.dtypes
```

```python
# creating a column for 'lead_time' to a 'lead_time_binned' column
data['lead_time'].unique
data['lead_time'].describe()

bins=[0,100,200,300,400,500,600,700,800]
labels=['0-100','101-200','201-300','301-400','401-500','501-600','601-700','701-800']
data['lead_time_binned']=pd.cut(data['lead_time'],bins=bins,labels=labels)
data[['lead_time','lead_time_binned']]
data['lead_time_binned'].value_counts()
```
